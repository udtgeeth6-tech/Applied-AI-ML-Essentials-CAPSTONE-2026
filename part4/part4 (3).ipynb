(A) Structured JSON Extraction: Given a dataset record formatted as a JSON object (with all relevant feature fields), use an LLM API call to extract or reformat specific fields into a target JSON schema. Define the target schema and validate the LLM output against it.
Objective

This task demonstrates how a Large Language Model (LLM) can be constrained to return data in a predefined JSON structure instead of free-form text. A dataset record is provided as a JSON object, and the LLM extracts key fields into a validated target schema.

```python
#reusable fn
import requests
import os

def call_llm(system_prompt,
             user_prompt,
             temperature=0.0,
             max_tokens=512):

    api_key = os.environ.get("LLM_API_KEY")

    if not api_key:
        print("LLM_API_KEY not found")
        return None

    url = "https://openrouter.ai/api/v1/chat/completions"

    payload = {
        "model": "openai/gpt-4o-mini",
        "messages": [
            {
                "role": "system",
                "content": system_prompt
            },
            {
                "role": "user",
                "content": user_prompt
            }
        ],
        "temperature": temperature,
        "max_tokens": max_tokens
    }

    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }

    response = requests.post(
        url,
        headers=headers,
        json=payload
    )

    if response.status_code != 200:
        print("Status Code:", response.status_code)
        print(response.text)
        return None

    return response.json()["choices"][0]["message"]["content"]
    #test
```

```python
import requests
import os

def call_llm(system_prompt, user_prompt, temperature=0.0, max_tokens=512):
    # Integrate PII guardrail here
    if 'has_pii' in globals() and has_pii(user_prompt):
        print("PII detected in user prompt. LLM call blocked.")
        return "PII detected. LLM call blocked."

    api_key = os.environ["LLM_API_KEY"]

    url = "https://openrouter.ai/api/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }

    payload = {
        "model": "openai/gpt-5.6-sol",
        "messages": [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        "temperature": temperature,
        "max_tokens": max_tokens
    }

    response = requests.post(
        url,
        headers=headers,
        json=payload
    )

    if response.status_code != 200:
        print("Status Code:", response.status_code)
        print(response.text)
        return None

    return response.json()["choices"][0]["message"]["content"]

    result = call_llm(
    "You are a helpful assistant.",
    "Reply with only the word hello"
    )

    print(result)
```

```python

#clean parsed
import json
from jsonschema import validate, ValidationError

# Expected schema
output_schema = {
    "type": "object",
    "properties": {
        "company": {"type": "string"},
        "cpu_brand": {"type": "string"},
        "ram_gb": {"type": "number"},
        "storage_summary": {"type": "string"},
        "price_category": {"type": "string"}
    },
    "required": [
        "company",
        "cpu_brand",
        "ram_gb",
        "storage_summary",
        "price_category"
    ]
}

# Fallback value
fallback = {
    "company": None,
    "cpu_brand": None,
    "ram_gb": None,
    "storage_summary": None,
    "price_category": None
}

# Example LLM output
llm_response = """
{
  "company": "HP",
  "cpu_brand": "Intel Core i3",
  "ram_gb": 8,
  "storage_summary": "128GB SSD + 500GB HDD",
  "price_category": "budget"
}
"""

try:
    # Parse JSON
    parsed_output = json.loads(llm_response)

    # Validate schema
    validate(instance=parsed_output, schema=output_schema)

    print("Validation successful!")
    print(parsed_output)

except ValidationError as e:
    print("Validation Error:")
    print(e.message)
    parsed_output = fallback

except json.JSONDecodeError as e:
    print("JSON Parsing Error:")
    print(e)
    parsed_output = fallback

print("\nFinal Output:")
print(parsed_output)
```
